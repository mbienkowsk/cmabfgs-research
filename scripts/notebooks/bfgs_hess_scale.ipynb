```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
RESULTS = "/home/maksym/studia/praca-inzynierska/thesis-2/scripts/bfgs_hess_scale.parquet"
df = pd.read_parquet(RESULTS)
df = df.drop(columns="run_id").interpolate(method="index")
df
```

```python
non_normalized = [col for col in df.columns if (not col.endswith("normalized") and not col.endswith("norm_probe"))]
print(non_normalized)
df.plot(title="Convergence of BFGS with different preconditioner inv hess matrices", y=non_normalized)
plt.yscale("log")
plt.show()
```

```python
normalized = ["best_inv_hess * 1"] + [col for col in df.columns if col.endswith("normalized")]
df.plot(title="Convergence of BFGS with different preconditioner inv hess matrices (with normalization)", y=normalized)
plt.yscale("log")
plt.show()
converged = [y for y in normalized if df.iloc[-1][y] < 1e5]
print(f"converged:\n{"\n".join(converged)}\n")

not_converged = [y for y in normalized if df.iloc[-1][y] > 1e5]
print(f"did not converge:\n{"\n".join(not_converged)}")
```

```python
normalized_probe = ["best_inv_hess * 1"] + [col for col in df.columns if col.endswith("norm_probe")]
df.plot(title="Convergence of BFGS with different preconditioner inv hess matrices (with normalization by probing)", y=normalized_probe)
plt.yscale("log")

converged = [y for y in normalized_probe if df.iloc[-1][y] < 1e5]
print(f"converged:\n{"\n".join(converged)}\n")

not_converged = [y for y in normalized_probe if df.iloc[-1][y] > 1e5]
print(f"did not converge:\n{"\n".join(not_converged)}")
plt.show()
```
